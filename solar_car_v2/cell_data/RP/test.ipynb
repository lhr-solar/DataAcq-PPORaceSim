```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
import numpy as np
import os
import json
import csv
```

```python
cd = os.getcwd()

data=[]

for filename in os.listdir(cd):
    if filename.endswith('.char'):
        filepath = os.path.join(cd, filename)
        with open(filepath, 'r') as file:
            data.append(json.load(file))
```

```python
df=pd.DataFrame(data[0]['points_orig'])
df
```

```python
maxes = []
for i in data:
    df=pd.DataFrame(i['points_orig'])
    voltages = np.array(df[0])
    currents = np.array(df[1])

    max_index = np.argmax(voltages * currents)
    voltage, current, power = voltages[max_index], currents[max_index], voltages[max_index]*currents[max_index]
    maxes.append([voltage, current, power])
```

```python
with open('IVCurve.csv', 'w') as f:
     
    # using csv.writer method from CSV package
    write = csv.writer(f)
     
    write.writerow(['Voltage', 'Current', 'Power'])
    write.writerows(maxes)
```
